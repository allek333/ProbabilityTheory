### Задача 1

Из колоды в 52 карты извлекаются случайным образом 4 карты.
1. Найти вероятность того, что все карты – крести. 
2. Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

Начнём решение последовательно. Для начало определим количество способов извлечь 4 карты из колоды в 52 карты. Для этого подойдёт формула для сочетаний. Физический смысл - извлекается случайная выборка и не важно в каком порядке. В математическом выражении будет:

Далее запишем это уравнение в кодах python.

```python
from math import factorial
```

```python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
combinations(52, 4)
```

Имеем 270.725 комбинаций, как можно выбрать 4 карты из 52.

#### Найти вероятность того, что все карты – крести.

Всего в колоде 13 треф. Количество способов достать 4 трефы из 13 составит:

В кодах питона это будет:

```python
combinations(13, 4)
```

Для нахождения вероятности разделим одно на другое: 715/270725 = 0.002 или 0.2%

```python
combinations(13, 4)/combinations(52, 4)
```

#### Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

В этой задаче условие на языке алгебры логики будет выглядеть: найти вероятность, что вытянут 1, 2, 3 или 4 туза. Значит это логическое ИЛИ, то есть сложение четырёх вероятностей. 

Или это будет выбор из остатка карт: "4 шанса из 52" или "3 из 51" и т.д.

не, не так

```python
(combinations(4, 4)+combinations(4, 3)+combinations(4, 2)+combinations(4, 1))/combinations(52, 4)
```

Ответ сравнимый. Примерно 0.006%
